### Question 1

```python
import pandas as pd
import numpy as np
```

```python
def answer_one():
    energy = pd.read_excel('Energy Indicators.xls', 
                        skiprows=17, 
                        skipfooter=38, 
                        usecols=[2, 3, 4, 5], 
                        names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'])

    # Numbering from 1
    energy.index = range(1, len(energy) + 1)

    # Round '% Renewable' to 2 decimal places
    energy['% Renewable'] = energy['% Renewable'].round(2)

    # Raplace '...' to NaN
    energy = energy.replace('...', np.nan)

    # Convert 'Energy Supply' to gigajoules
    energy['Energy Supply'] *= 1000000

    # Rename countries
    country_replace = {
        "Republic of Korea": "South Korea",
        "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"
    }
    energy['Country'] = energy['Country'].replace(country_replace)

    # Delete numbers and parentheses
    energy['Country'] = energy['Country'].str.replace(r'\d', '', regex=True)
    energy['Country'] = energy['Country'].str.replace(r'\([^)]*\)', '', regex=True)
    
    # Load GDP data from "world_bank.csv"
    GDP = pd.read_csv("world_bank.csv", skiprows=4)
    GDP.index = range(1, len(GDP) + 1)

    # Rename countries  
    country_replace = {
        "Korea, Rep.": "South Korea",
        "Iran, Islamic Rep.": "Iran",
        "Hong Kong SAR, China": "Hong Kong"
    }
    GDP['Country Name'] = GDP['Country Name'].replace(country_replace)

    ScimEn = pd.read_excel('Energy Engineering and Power Technology.xlsx',)
    ScimEn.index = range(1, len(ScimEn) + 1)

    # Merge Energy and ScimEn
    energy_SciemEn_merged = pd.merge(energy, ScimEn, how='inner', on='Country')

    # Filter the top 15 countries by 'Rank'
    top_15_countries = energy_SciemEn_merged[energy_SciemEn_merged['Rank'] <= 15]

    # Extract GDP data for the years 2006-2015
    get_GDP = ['Country Name', 
                '2006', '2007', 
                '2008', '2009', 
                '2010', '2011', 
                '2012', '2013', 
                '2014', '2015']
    GDP = GDP[get_GDP]

    # Merge the filtered top 15 countries with GDP data
    energy_SciemEn_GDP_merged = pd.merge(top_15_countries, GDP, how='inner', left_on='Country', right_on='Country Name')

    # Drop the 'Country Name'
    energy_SciemEn_GDP_merged.drop('Country Name', axis=1, inplace=True)

    # Set index
    energy_SciemEn_GDP_merged.set_index('Country', inplace=True)

    # Need cloumns
    columns_need = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document',
                    'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable',
                    '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']

    energy_SciemEn_GDP_merged = energy_SciemEn_GDP_merged[columns_need]
    
    return energy_SciemEn_GDP_merged
```

```python
Result = answer_one()
Result
```

### Question 2

```python
def answer_two():
    Top15 = answer_one()
    Top15['Average GDP'] = Top15[['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']].mean(axis=1)

    Top15.sort_values(by='Average GDP', ascending=False, inplace=True)
    
    average_GDP = Top15['Average GDP']

    return average_GDP
```

```python
Result = answer_two()
Result
```

```python
def answer_three():
    Top15 = answer_one()
    Top6 = Top15.iloc[5]
    
    GDP_2006 = Top6['2006']
    GDP_2015 = Top6['2015']

    GDP_change = GDP_2015 - GDP_2006
    Percentage_change = (GDP_change / GDP_2006) * 100

    return Percentage_change
```

```python
def answer_four():
    Top15 = answer_one()
    Top15['Citations Ratio'] = Top15['Self-citations'] / Top15['Citations']

    max_ratio_country = Top15['Citations Ratio'].idxmax()
    max_ratio_value = Top15['Citations Ratio'].max()

    return max_ratio_country, max_ratio_value
```

```python
def answer_six():
    Top15 = answer_one()
    Top15['Citable documents per Capita'] = Top15['Citable documents'] / Top15['Energy Supply per Capita']
    
    correlation = Top15['Citable documents per Capita'].corr(Top15['Energy Supply per Capita'])
    
    return correlation
```

### Question 7

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict = {
        'China': 'Asia',
        'United States': 'North America',
        'Japan': 'Asia',
        'United Kingdom': 'Europe',
        'Russian Federation': 'Europe',
        'Canada': 'North America',
        'Germany': 'Europe',
        'India': 'Asia',
        'France': 'Europe',
        'South Korea': 'Asia',
        'Italy': 'Europe',
        'Spain': 'Europe',
        'Iran': 'Asia',
        'Australia': 'Australia',
        'Brazil': 'South America'
    }

    Continents_df = pd.DataFrame(columns=['size', 'sum', 'mean', 'std'])

    for continent, countries in ContinentDict.items():
        population = Top15[Top15.index.isin([country for country, cont in ContinentDict.items() if cont == countries])]['Energy Supply']
        Continents_df.loc[countries] = [len(population), population.sum(), population.mean(), population.std()]

    return Continents_df
```

```python
Result = answer_seven()
Result
```
