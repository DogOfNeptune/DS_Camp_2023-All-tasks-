# Use sklearn.datasets.load_breast_cancer and apply sklearn.LogisticRegression

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
import matplotlib.patches as mpatches
%matplotlib inline
```

### Load data set

```python
from sklearn.datasets import load_breast_cancer
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

print('Accuracy on the training subset: {:.3f}'.format(logreg.score(X_train, y_train)))
print('Accuracy on the test subset: {:.3f}'.format(logreg.score(X_test, y_test)))
```

### Visualize the results

```python
plt.figure(figsize=(8,6))
plt.title('Breast Cancer Logistic Regression', fontsize=18)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Target', fontsize=14)

feature_names = load_breast_cancer().feature_names
coef = logreg.coef_[0]
colors = ['red' if c < 0 else 'blue' for c in coef]
plt.bar(feature_names, coef, color=colors)
plt.xticks(rotation=85)
plt.show() 
```
