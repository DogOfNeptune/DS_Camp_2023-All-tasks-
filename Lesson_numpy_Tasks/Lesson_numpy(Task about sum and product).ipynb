# Task about sum and product

There are two whole numbers:
1 < a,b <100

One scientist("Sum") get provided with sum of numbers,
another  ("Prod") get provided with product of numbers. 
Both scientists know that numbers 1 < a,b <100.

Determine the numbers being based on the following dialog: 
    Prod: I don't know the numbers;
    Sum: I know it;
    Prod: then I know the numbers; 
    Sum: then I know the numbers too.

```python
import numpy as np
```

# Prod: I don't know the numbers

Prod, дізнавшись добуток двох цілих чисел, більших за одиницю, розкладе його на множники, враховуючи, що існує більше одного способу розкладання.

Sum, дізнавшись суму, візьмеся розкладати її на доданки, забезпечуючи тим самим те, що будь-яке розбиття на доданки, більші за одиницю, має добуток доданків, що розкладається на множники більш ніж одним способом.

Щоб це забезпечити, важливо, щоб сума, відома Sum, не представлялася у вигляді суми двох простих чисел.

Також сума не повина перевищувати 198, оскільки 99 + 99 = 198.

```python
def prime_numb(number):
    if number % 2 == 0:
        return number == 2
    d = 3
    while d * d <= number and number % d != 0:
        d += 2
    return d * d > number
        
result = prime_numb(11)
result
```

```python
def split_to_pairs_sum(number):
    sum_combinations = np.array([]).reshape(0, 2)
    for i in range(2, number // 2 + 1):
        sum_combinations = np.append(sum_combinations, [(i, number - i)], axis=0)
        
    return sum_combinations

result = split_to_pairs_sum(11)
result
```

```python
def After_first_phrase():
    all_sums = np.arange(4, 198)
    need_sums = np.copy(all_sums)
    
    for sum in all_sums:
        possible_all_pairs_of_one_sum = split_to_pairs_sum(sum)
        
        for pair in possible_all_pairs_of_one_sum:
            if prime_numb(pair[0]) and prime_numb(pair[1]):
                indices_to_remove = np.where(need_sums == sum)
                need_sums = np.delete(need_sums, indices_to_remove)
                break
    
    return need_sums

result_1 = After_first_phrase()
result_1
```

# Sum: I know it

Перевіряєм доданки кожної суми. З кожної пари доданків ми повинні скласти добуток. Потім розкласти цей добуток на доданки якщо вже є сума однієї з пар є в списку сум, то не додаєм їх, якщо немає то привязуємо її до суми. Також не забуваєм, що не може бути два числа простих.

```python
def multiplier_pairs(number):
    pairs = []
    
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            pairs.append([i, number // i])
    return pairs

result = multiplier_pairs(18)
result
```

```python
def After_second_phrase(all_sums_after_delete):
    def is_valid_pair(pair_terms_sum_int, all_multipliers_one_pair):
        for one_pair in all_multipliers_one_pair:
            one_pair_int = [int(x) for x in one_pair]
            if prime_numb(one_pair[0]) and prime_numb(one_pair[1]):
                return False
            if sum(one_pair) in all_sums_after_delete and one_pair_int != pair_terms_sum_int:
                return False
        return True

    potential_answers = {}
    
    for sum_value in all_sums_after_delete:
        all_pairs_terms_sum = split_to_pairs_sum(sum_value)
        potential_answers[sum_value] = []
        
        for pair_terms_sum in all_pairs_terms_sum:
            pair_terms_sum_int = [int(x) for x in pair_terms_sum]
            all_multipliers_one_pair = multiplier_pairs(pair_terms_sum_int[0] * pair_terms_sum_int[1])
            permit = is_valid_pair(pair_terms_sum_int, all_multipliers_one_pair)
            
            if permit:
                potential_answers[sum_value].append(pair_terms_sum_int)

    return potential_answers

result_2 = After_second_phrase(result_1)
result_2
    
```

# Prod: then I know the numbers

Почувши репліку Sum, Prod міг упевнено заявити, що знає задумані два числа, оскільки тільки одне з його відомих розкладень на множники відповідало сумі, яка відповідала другому вислову.

```python
def After_third_phrases(possible_answers):
    return [values for values in possible_answers.values() if len(values) == 1]
          
result_3 = After_third_phrases(result_2)
result_3
```

# Sum: then I know the numbers too.

Аналогічна ситуація з реплікою Prod

```python
def After_four_phrases(possible_answers):
    return [values for values in possible_answers.values() if len(values) == 1]

result_4 = After_third_phrases(result_2)
result_4
```
