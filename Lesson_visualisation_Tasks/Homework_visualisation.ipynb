```python
import pandas as pd
import numpy as np
```

```python
Car_insurance_claim = pd.read_csv('Car_Insurance_Claim.csv', index_col=0)
Car_insurance_claim
```

# Data Cleaning

```python
Car_insurance_claim.isna().sum()  
```

```python
Car_insurance_claim.dropna(inplace=True)
Car_insurance_claim.isna().sum()
```

```python
count_of_none = (Car_insurance_claim == 'none').sum()
count_of_none
```

```python
rows_with_none = Car_insurance_claim[Car_insurance_claim == 'none'].any(axis=1)

Car_insurance_claim = Car_insurance_claim[~rows_with_none]
count_of_none = (Car_insurance_claim == 'none').sum()
count_of_none
```

```python
Car_insurance_claim
```

# Descriptive Statistics

```python
Car_insurance_claim.dtypes
```

```python
Car_insurance_claim = Car_insurance_claim.astype(str)

Car_insurance_claim['CREDIT_SCORE'] = Car_insurance_claim['CREDIT_SCORE'].astype(float)
Car_insurance_claim['VEHICLE_OWNERSHIP'] = Car_insurance_claim['VEHICLE_OWNERSHIP'].astype(float).astype(int)
Car_insurance_claim['MARRIED'] = Car_insurance_claim['MARRIED'].astype(float).astype(int)
Car_insurance_claim['CHILDREN'] = Car_insurance_claim['CHILDREN'].astype(float).astype(int)
Car_insurance_claim['ANNUAL_MILEAGE'] = Car_insurance_claim['ANNUAL_MILEAGE'].astype(float)
Car_insurance_claim['SPEEDING_VIOLATIONS'] = Car_insurance_claim['SPEEDING_VIOLATIONS'].astype(int)
Car_insurance_claim['DUIS'] = Car_insurance_claim['DUIS'].astype(int)
Car_insurance_claim['PAST_ACCIDENTS'] = Car_insurance_claim['PAST_ACCIDENTS'].astype(int)

Car_insurance_claim_str = Car_insurance_claim.select_dtypes(include='object')
Car_insurance_claim_str = Car_insurance_claim_str.drop('ID', axis=1).set_index('OUTCOME')
Car_insurance_claim_str.describe()
```

```python
Car_insurance_claim['OUTCOME'] = pd.to_numeric(Car_insurance_claim['OUTCOME'], errors='coerce')

Car_insurance_claim_num = Car_insurance_claim.select_dtypes(include=[np.number])
Car_insurance_claim_num = Car_insurance_claim_num.set_index('OUTCOME')
Car_insurance_claim_num.describe()
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid", palette="pastel", color_codes=True)
plt.figure(figsize=(20, 20))

for i, column in enumerate(Car_insurance_claim_num.columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data=Car_insurance_claim_num, x=column, kde=True)
    plt.title(f'{column} VS OUTCOME', fontsize=16)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)

# Зроби ще графік для індексу OUTCOME
plt.subplot(3, 3, 9)
sns.histplot(data=Car_insurance_claim, x='OUTCOME', kde=True)

plt.tight_layout()
plt.show()
```

### Conclusions:
- People with higher credit scores have a greater chance of a positive outcome.
- People who own a car are 2.5 times more likely to receive insurance payments. 
- Married people are about 10% more likely to receive insurance payments.
- People with children are 50% more likely to receive insurance payments. 
- In general, about 40% of people receive insurance payments

```python
string_columns = ['AGE', 'GENDER', 'RACE', 'DRIVING_EXPERIENCE', 'EDUCATION', 'INCOME', 'VEHICLE_YEAR', 'POSTAL_CODE', 'VEHICLE_TYPE']

sns.set(style="darkgrid", palette="pastel", color_codes=True)
plt.figure(figsize=(20, 20))

for i, column in enumerate(string_columns, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=Car_insurance_claim, x=column, hue='OUTCOME')
    plt.title(f'{column} vs OUTCOME', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(title='OUTCOME', title_fontsize=10)

plt.tight_layout()
plt.show()
```

### Conclusions:
- The average percentage of people receiving compensation is 40%.
- People aged 16 to 25 have the highest percentage of compensation.
- People aged 40 to 64 years have the lowest percentage of compensation.
- Women have a lower percentage of compensation than men.
- Majority people have a higher percentage of receiving compensation than people of other races.
- Drivers with 10 to 19 years of experience are more likely to receive insurance indemnity than those without - more than half of them.
- Drivers with higher or incomplete higher education do not differ much from each other, but people with higher education receive more insurance indemnity.
- Drivers with higher incomes often do not receive insurance payments compared to other classes: the middle class receives 25%, the poor 70%, and the working class 50%.
- Vehicles manufactured before 2015 are more likely to receive an insurance claim
- Vehicles registered in certain postal codes are more likely to receive an insurance claim, most often 32765, 92101, 21217.
- In general, both sports cars and sedans have approximately the same percentage of insurance claims, just a different number of them in total.
