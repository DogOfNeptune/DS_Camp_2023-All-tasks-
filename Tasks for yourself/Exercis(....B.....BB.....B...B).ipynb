# Task about '.....B.....BB.....B...B' -> '.....B.B.B.B.B.........'

### Initialisation string

```python
import random

def get_random_string():
    result_string = ''.join(random.choices(['B', '.'], k=10000))
    return result_string

string = get_random_string()
string
```

### Algorithm

```python
def coun_min_swaps_B(string):
    count_B = string.count('B')
    count_dot = string.count('.')
    if count_B >= count_dot + 2:
        return string, -1

    first_B = string.index('B')
    
    if first_B % 2 == 0:
        range_B_indices = range(0, len(string), 2)
    else:
        range_B_indices = range(1, len(string), 2)
    
    all_B = [B for B in string if B == 'B']
    new_string = len(string) * ['.']
    for i, B in zip(range_B_indices, all_B):
        new_string[i] = B
    
    if new_string == string:
        return string, 0
    
    original_B_indices = [i for i, x in enumerate(string) if x == 'B']
    sorted_B_indices = [i for i, x in enumerate(new_string) if x == 'B']
    correct_positions = len(set(original_B_indices).intersection(sorted_B_indices))
    min_swaps = len(all_B) - correct_positions
    
    return new_string, min_swaps
```

```python
for i in range(5):
    string = get_random_string()
    print('Generated string: ', string)
    
    string, swaps = coun_min_swaps_B(string)
    
    print('Reworked string: ', string)
    print('Min amount of swaps: ', swaps)
    print()
```
