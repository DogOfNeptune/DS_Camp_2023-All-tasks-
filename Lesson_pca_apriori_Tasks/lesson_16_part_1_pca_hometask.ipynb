### Load Data

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
```

```python
X, y = load_breast_cancer(return_X_y= True)
```

### Scale

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

### PCA

```python
from sklearn.decomposition import PCA 
```

```python
# This function checks how much data was discarded
def comress_score(X, X_approximate):
    return np.sum(np.apply_along_axis (np.linalg.norm,1, (X-X_approximate))**2)/\
        np.sum(np.apply_along_axis (np.linalg.norm,1, X)**2)
```

```python
pca = PCA(n_components=10)
Z = pca.fit_transform(X_scaled)
print ('Z.shape = ',Z.shape)

X_approximate = pca.inverse_transform(Z)
# Compress score = 0.048 it is not bad score. It must be to 0.05, but this score for 10 components.
print ('Compress score= {0:.3f}'.format(comress_score(X_scaled, X_approximate)))
```

### Visualisation 2 principle components.

```python
# Visualize 2 principle components on the scatter
plt.scatter(Z[:,0], Z[:,1], c=y, edgecolor='black', label='negative', alpha=0.8)
plt.title('PCA (2 components) for breast cancer dataset')
plt.xlabel('First principle components')
plt.ylabel('Second principle components')
```
